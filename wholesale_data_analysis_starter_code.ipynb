## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['line_subtotal']=df['unit_price']*df['qty']
df[['unit_price','qty','line_subtotal']]
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['total_weight']=df['unit_weight']*df['qty']

# Define a function
def shipping_price(total_weight):
    if total_weight < 50:
        return total_weight * 10
    return total_weight * 7

# "Apply" the function to the shipping_price column



df['shipping_price']=df['total_weight'].apply(shipping_price)
df[['unit_price','unit_weight','qty','total_weight','shipping_price']]
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['line_price']=(df['line_subtotal']+df['shipping_price'])*(1+0.0925)
df[['line_subtotal','shipping_price','line_price']]
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

df['line_cost']=(df['unit_cost']* df['qty'])+df['shipping_price']
df
```

```python
# Create a column for the profit of each line using line cost and line price
df['line_profit']=df['line_price']-df['line_cost']
df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

Client_price=df.groupby(['client_id'])['line_price'].sum()

for clients in top_clients_by_quantity:
   print(f"Client ID {clients}: ${Client_price[clients]:,.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
pd.options.display.float_format = '{:,.2f}'.format
df_topClients=df[df['client_id'].isin(top_clients_by_quantity)]

df_topClients_sum= df_topClients.groupby(['client_id'])[['qty','shipping_price','line_price','line_cost','line_profit']].sum()
df_topClients_sum=df_topClients_sum.reset_index()
df_topClients_sum
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
col={'client_id':'Client ID',
     'qty':'Units',
    'shipping_price':'Shipping',
    'line_price':'Total Revenue',
     'line_cost':'Total Cost',
     'line_profit':'Total Profit' 
}
df_topClients_sum=df_topClients_sum.rename(columns=col)
```

```python
# Define a function that converts a dollar amount to millions.

def currency_format_millions(amount):
    return (amount/1000000)
```

```python
# Apply the currency_format_millions function to only the money columns.


# To ensure that original dataframe is returned if program is re-run a  reset function is used  that returns the original DataFrame
def reset_dataframe(original_df):
    return original_df.copy()

# Create a copy of the original DataFrame for resetting
original_df_topClients_sum = df_topClients_sum.copy()

# Reset the DataFrame before applying the conversion
df_topClients_sum_mil = reset_dataframe(original_df_topClients_sum)


df_topClients_sum_mil['Shipping']=df_topClients_sum_mil['Shipping'].apply(currency_format_millions)
df_topClients_sum_mil['Total Revenue']=df_topClients_sum_mil['Total Revenue'].apply(currency_format_millions)
df_topClients_sum_mil['Total Cost']=df_topClients_sum_mil['Total Cost'].apply(currency_format_millions)
df_topClients_sum_mil['Total Profit']=df_topClients_sum_mil['Total Profit'].apply(currency_format_millions)

```

```python

# Rename the columns to reflect the change in the money format. 

col_mil={'Shipping':'Shipping (millions)',
    'Total Revenue': 'Total Revenue (millions)',
     'Total Cost': 'Total Cost(millions) ',
    'Total Profit' : 'Total Profit (millions)'
}
df_topClients_summary_millions=df_topClients_sum_mil.rename(columns=col_mil)

df_topClients_summary_millions
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
df_topClients_summary_millions_sort=df_topClients_summary_millions.sort_values(by="Total Profit (millions)", ascending=False)
df_topClients_summary_millions_sort
```
